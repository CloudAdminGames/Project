## 2Ô∏è‚É£ Analyse des valeurs manquantes

```python
# Calcul des valeurs manquantes par colonne
missing_data = pd.DataFrame({
    'Colonne': df.columns,
    'Valeurs manquantes': df.isnull().sum().values,
    'Pourcentage (%)': (df.isnull().sum().values / len(df) * 100).round(2),
    'Type': df.dtypes.values
})
missing_data = missing_data.sort_values('Pourcentage (%)', ascending=False)
print("üìâ Colonnes avec le plus de valeurs manquantes :")
missing_data[missing_data['Pourcentage (%)'] > 0]
```

```python
# V√©rifier aussi les valeurs "nan" en texte (souvent pr√©sentes dans les CSV)
nan_text_counts = {}
for col in df.columns:
    if df[col].dtype == 'object':
        count = df[col].astype(str).str.lower().isin(['nan', 'none', '', 'null']).sum()
        if count > 0:
            nan_text_counts[col] = count

print("‚ö†Ô∏è Colonnes avec des valeurs 'nan'/'none'/vides en texte :")
for col, count in sorted(nan_text_counts.items(), key=lambda x: x[1], reverse=True):
    pct = count / len(df) * 100
    print(f"  {col}: {count:,} ({pct:.1f}%)")
```

## 7Ô∏è‚É£ R√©sum√© et recommandations de nettoyage

```python
print("=" * 60)
print("üìã R√âSUM√â DU NETTOYAGE RECOMMAND√â")
print("=" * 60)

# Calculer les crit√®res de nettoyage
total = len(df)

# Crit√®re 1: Sans rating ET sans votes
crit1 = df[(df['rating'].isna() | (df['rating'] == 0)) & 
           (df['ratings_count'].isna() | (df['ratings_count'] == 0))]
print(f"\n1Ô∏è‚É£ Sans rating ET sans votes : {len(crit1):,} jeux ({len(crit1)/total*100:.1f}%)")

# Crit√®re 2: Sans date de sortie
crit2 = df[df['released_date'].isna()]
print(f"2Ô∏è‚É£ Sans date de sortie : {len(crit2):,} jeux ({len(crit2)/total*100:.1f}%)")

# Crit√®re 3: Sans description
if 'description' in df.columns:
    crit3 = df[(df['description'].isna()) | (df['description'].astype(str).str.len() < 10)]
    print(f"3Ô∏è‚É£ Sans description : {len(crit3):,} jeux ({len(crit3)/total*100:.1f}%)")

# Crit√®re combin√© sugg√©r√© (jeux "pauvres en donn√©es")
poor_data = df[
    ((df['rating'].isna()) | (df['rating'] == 0)) &
    ((df['ratings_count'].isna()) | (df['ratings_count'] == 0)) &
    (df['released_date'].isna())
]
print(f"\nüóëÔ∏è Jeux tr√®s incomplets (sans rating, votes ET date) : {len(poor_data):,} jeux")
```

## 8Ô∏è‚É£ Appliquer le nettoyage (optionnel)

```python
# Cr√©er un dataset nettoy√© avec diff√©rents niveaux de filtrage

# Niveau 1: Supprimer uniquement les jeux sans AUCUNE √©valuation
df_clean_light = df[
    ~((df['rating'].isna() | (df['rating'] == 0)) & 
      (df['ratings_count'].isna() | (df['ratings_count'] == 0)))
].copy()
print(f"üì¶ Nettoyage l√©ger : {len(df_clean_light):,} jeux restants ({len(df_clean_light)/total*100:.1f}%)")

# Niveau 2: Supprimer jeux sans rating ET sans date
df_clean_medium = df[
    ~((df['rating'].isna() | (df['rating'] == 0)) & df['released_date'].isna())
].copy()
print(f"üì¶ Nettoyage moyen : {len(df_clean_medium):,} jeux restants ({len(df_clean_medium)/total*100:.1f}%)")

# Niveau 3: Garder uniquement les jeux avec au moins 5 votes
df_clean_strict = df[df['ratings_count'] >= 5].copy()
print(f"üì¶ Nettoyage strict (5+ votes) : {len(df_clean_strict):,} jeux restants ({len(df_clean_strict)/total*100:.1f}%)")
```

## üóëÔ∏è Analyse des colonnes √† supprimer

```python
# Analyse d√©taill√©e de chaque colonne pour d√©cider lesquelles supprimer
print("=" * 80)
print("üìã ANALYSE D√âTAILL√âE DES COLONNES")
print("=" * 80)

# Calculer le % de valeurs vides/nan pour chaque colonne
col_analysis = []
for col in df.columns:
    null_count = df[col].isna().sum()
    # Compter aussi les 'nan' en texte
    if df[col].dtype == 'object':
        nan_text = df[col].astype(str).str.lower().isin(['nan', 'none', '', 'null']).sum()
    else:
        nan_text = 0
    
    total_empty = max(null_count, nan_text)
    pct_empty = total_empty / len(df) * 100
    
    col_analysis.append({
        'Colonne': col,
        'Type': str(df[col].dtype),
        'Vides (%)': round(pct_empty, 1),
        'Valeurs uniques': df[col].nunique()
    })

col_df = pd.DataFrame(col_analysis).sort_values('Vides (%)', ascending=False)
print("\nüìä Colonnes tri√©es par % de valeurs vides :\n")
print(col_df.to_string(index=False))
```

```python
# Propositions de colonnes √† SUPPRIMER
print("=" * 80)
print("üóëÔ∏è COLONNES PROPOS√âES POUR SUPPRESSION")
print("=" * 80)

# Cat√©gorie 1: >90% vides
cols_very_empty = col_df[col_df['Vides (%)'] >= 90]['Colonne'].tolist()
print(f"\nüî¥ CAT√âGORIE 1 - Plus de 90% vides ({len(cols_very_empty)} colonnes):")
for c in cols_very_empty:
    pct = col_df[col_df['Colonne'] == c]['Vides (%)'].values[0]
    print(f"   ‚Ä¢ {c} ({pct}% vide)")

# Cat√©gorie 2: Colonnes redondantes ou inutiles pour une API
cols_useless = [
    'saturated_color',      # Couleur technique de l'image
    'dominant_color',       # Couleur technique de l'image
    'suggestions_count',    # Pas utile pour l'API
    'updated',              # Date de mise √† jour RAWG, pas pertinent
    'reviews_text_count',   # Redondant avec reviews_count
    'added',                # Nombre d'ajouts sur RAWG
    'clip',                 # Clips vid√©o (souvent vide)
    'parents_count',        # Li√© aux DLC/versions
    'creators_count',       # Peu pertinent
    'twitch_count',         # Stats Twitch obsol√®tes
    'youtube_count',        # Stats YouTube obsol√®tes  
    'reddit_count',         # Stats Reddit
    'screenshots_count',    # On a d√©j√† short_screenshots
    'parent_achievements_count',  # Li√© aux achievements parents
    'game_series_count',    # Peut √™tre calcul√©
    'additions_count',      # DLC count
    'movies_count',         # Peu utile
    'tba',                  # "To be announced" - peu utile
]

print(f"\nüü† CAT√âGORIE 2 - Colonnes non pertinentes pour l'API ({len(cols_useless)} colonnes):")
for c in cols_useless:
    if c in df.columns:
        print(f"   ‚Ä¢ {c}")

# Cat√©gorie 3: Colonnes dupliqu√©es
cols_duplicate = [
    'description',          # On garde description_raw (plus complet)
    'name_original',        # Redondant avec name dans 99% des cas
]
print(f"\nüü° CAT√âGORIE 3 - Colonnes redondantes ({len(cols_duplicate)} colonnes):")
for c in cols_duplicate:
    if c in df.columns:
        print(f"   ‚Ä¢ {c}")

# Total
all_to_remove = list(set(cols_very_empty + cols_useless + cols_duplicate))
all_to_remove = [c for c in all_to_remove if c in df.columns]
print(f"\n" + "=" * 80)
print(f"üìä TOTAL: {len(all_to_remove)} colonnes √† supprimer sur {len(df.columns)}")
print(f"üìä Colonnes restantes: {len(df.columns) - len(all_to_remove)}")
print("=" * 80)
```

```python
# Afficher les colonnes qui seront CONSERV√âES
cols_to_keep = [c for c in df.columns if c not in all_to_remove]
print("‚úÖ COLONNES √Ä CONSERVER POUR L'API :")
print("-" * 50)
for i, c in enumerate(cols_to_keep, 1):
    pct_empty = col_df[col_df['Colonne'] == c]['Vides (%)'].values[0]
    print(f"{i:2d}. {c:<30} ({pct_empty}% vide)")
```

## ‚úÖ NETTOYAGE FINAL ET EXPORT

```python
# =============================================================================
# NETTOYAGE COMPLET DU DATASET
# =============================================================================

print("üöÄ D√©but du nettoyage...")
print(f"üìä Dataset original : {len(df):,} jeux, {len(df.columns)} colonnes")

# 1. Colonnes √† supprimer (valid√©es)
cols_to_remove = [
    # >90% vides
    'reddit_name', 'reddit_logo', 'reddit_description', 'reddit_url', 'reddit_count',
    'metacritic_platforms', 'metacritic_url', 'metacritic', 'clip', 'alternative_names',
    'reactions', 'ratings', 'esrb_rating', 'rating', 'rating_top', 'ratings_count',
    'reviews_count', 'reviews_text_count', 'playtime', 'achievements_count',
    'parent_achievements_count', 'twitch_count', 'youtube_count', 'movies_count',
    'creators_count', 'tba', 'parents_count', 'additions_count', 'game_series_count',
    # Non pertinentes
    'saturated_color', 'dominant_color', 'suggestions_count', 'updated', 'added',
    'screenshots_count',
    # Redondantes
    'description', 'name_original',
    # Ajouts sugg√©r√©s
    'added_by_status', 'website', 'released_date', 'year'
]

# Filtrer uniquement les colonnes qui existent
cols_to_remove = [c for c in cols_to_remove if c in df.columns]

# 2. Cr√©er le dataset nettoy√©
df_clean = df.copy()

# Supprimer les colonnes
df_clean = df_clean.drop(columns=cols_to_remove)
print(f"‚úÇÔ∏è  Colonnes supprim√©es : {len(cols_to_remove)}")

# 3. Filtrer les jeux (garder ceux avec au moins 1 vote = nettoyage l√©ger)
# On utilise le df original pour le filtre car on a supprim√© ratings_count
df_clean = df_clean[df['ratings_count'].notna() & (df['ratings_count'] > 0)]
print(f"üéÆ Jeux avec au moins 1 vote conserv√©s")

# 4. Supprimer les doublons par slug
before_dedup = len(df_clean)
df_clean = df_clean.drop_duplicates(subset=['slug'], keep='first')
print(f"üîÑ Doublons supprim√©s : {before_dedup - len(df_clean)}")

# 5. Nettoyer les valeurs 'nan' en texte
for col in df_clean.columns:
    if df_clean[col].dtype == 'object':
        df_clean[col] = df_clean[col].replace(['nan', 'NaN', 'None', 'none', ''], np.nan)

print(f"\n‚úÖ Dataset nettoy√© : {len(df_clean):,} jeux, {len(df_clean.columns)} colonnes")
print(f"üìâ R√©duction : {(1 - len(df_clean)/len(df))*100:.1f}% des lignes supprim√©es")
```

```python
# Aper√ßu du dataset nettoy√©
print("üìã Colonnes finales :")
print(df_clean.columns.tolist())
print(f"\nüìä Aper√ßu des donn√©es :")
df_clean.head()
```

```python
# Sauvegarder le dataset nettoy√©
output_file = 'rawg-games-cleaned.csv'
df_clean.to_csv(output_file, index=False)
print(f"üíæ Dataset sauvegard√© : {output_file}")
print(f"üì¶ Taille : {len(df_clean):,} jeux √ó {len(df_clean.columns)} colonnes")
```

## üîç ANALYSE CORRECTE : Colonnes nulles APR√àS filtrage des jeux

```python
# =============================================================================
# BONNE M√âTHODE : Filtrer d'abord les jeux, puis analyser les colonnes
# =============================================================================

print("üîç ANALYSE CORRECTE DES COLONNES")
print("=" * 80)

# 1. D'abord filtrer les jeux (garder ceux avec au moins 1 vote)
df_filtered = df[df['ratings_count'].notna() & (df['ratings_count'] > 0)].copy()
print(f"üìä Jeux avec au moins 1 vote : {len(df_filtered):,} (sur {len(df):,})")

# 2. Analyser les colonnes APR√àS filtrage
print(f"\nüìã Analyse des colonnes sur les {len(df_filtered):,} jeux filtr√©s :")
print("-" * 80)

col_analysis_filtered = []
for col in df_filtered.columns:
    null_count = df_filtered[col].isna().sum()
    # Compter aussi les 'nan' en texte
    if df_filtered[col].dtype == 'object':
        nan_text = df_filtered[col].astype(str).str.lower().isin(['nan', 'none', '', 'null']).sum()
    else:
        nan_text = 0
    
    total_empty = max(null_count, nan_text)
    pct_empty = total_empty / len(df_filtered) * 100
    
    col_analysis_filtered.append({
        'Colonne': col,
        'Type': str(df_filtered[col].dtype),
        'Vides (%)': round(pct_empty, 1),
        'Valeurs uniques': df_filtered[col].nunique()
    })

col_df_filtered = pd.DataFrame(col_analysis_filtered).sort_values('Vides (%)', ascending=False)
print(col_df_filtered.to_string(index=False))
```

```python
# Comparer AVANT vs APR√àS filtrage
print("=" * 80)
print("üìä COMPARAISON : % vide AVANT vs APR√àS filtrage des jeux")
print("=" * 80)

# Fusionner les deux analyses
comparison = col_df[['Colonne', 'Vides (%)']].merge(
    col_df_filtered[['Colonne', 'Vides (%)']], 
    on='Colonne', 
    suffixes=('_AVANT', '_APR√àS')
)
comparison['Diff√©rence'] = comparison['Vides (%)_AVANT'] - comparison['Vides (%)_APR√àS']
comparison = comparison.sort_values('Diff√©rence', ascending=False)

print("\nüîÑ Colonnes avec la plus GRANDE diff√©rence (am√©lior√©es apr√®s filtrage) :")
print(comparison[comparison['Diff√©rence'] > 10].to_string(index=False))

print("\n" + "=" * 80)
print("‚ö†Ô∏è COLONNES √Ä R√â√âVALUER (√©taient >90% vides, maintenant potentiellement utiles) :")
print("=" * 80)
reval = comparison[(comparison['Vides (%)_AVANT'] >= 90) & (comparison['Vides (%)_APR√àS'] < 50)]
if len(reval) > 0:
    print(reval.to_string(index=False))
else:
    print("Aucune colonne √† r√©√©valuer")
```

```python
# Nouvelle proposition de colonnes √† supprimer bas√©e sur l'analyse APR√àS filtrage
print("=" * 80)
print("üóëÔ∏è NOUVELLE PROPOSITION DE COLONNES √Ä SUPPRIMER (bas√©e sur donn√©es filtr√©es)")
print("=" * 80)

# Colonnes >80% vides APR√àS filtrage
cols_empty_after = col_df_filtered[col_df_filtered['Vides (%)'] >= 80]['Colonne'].tolist()
print(f"\nüî¥ >80% vides APR√àS filtrage ({len(cols_empty_after)} colonnes):")
for c in cols_empty_after:
    pct_before = col_df[col_df['Colonne'] == c]['Vides (%)'].values[0]
    pct_after = col_df_filtered[col_df_filtered['Colonne'] == c]['Vides (%)'].values[0]
    print(f"   ‚Ä¢ {c}: {pct_before}% ‚Üí {pct_after}%")

# Colonnes non pertinentes (ind√©pendamment du % vide)
cols_not_useful = [
    'saturated_color', 'dominant_color',  # Couleurs techniques
    'suggestions_count', 'updated', 'added',  # Stats RAWG internes
    'screenshots_count',  # Redondant avec short_screenshots
    'description',  # On garde description_raw
    'name_original',  # Redondant avec name
]
print(f"\nüü† Non pertinentes pour l'API ({len(cols_not_useful)} colonnes):")
for c in cols_not_useful:
    print(f"   ‚Ä¢ {c}")

# Colonnes d√©riv√©es (on peut les recr√©er)
cols_derived = ['released_date', 'year']
print(f"\nüü° Colonnes d√©riv√©es √† supprimer ({len(cols_derived)} colonnes):")
for c in cols_derived:
    print(f"   ‚Ä¢ {c}")

# COLONNES √Ä GARDER (qui √©taient jug√©es vides mais sont maintenant utiles)
print("\n" + "=" * 80)
print("‚úÖ COLONNES √Ä CONSERVER (utiles apr√®s filtrage) :")
print("=" * 80)

cols_to_keep_new = col_df_filtered[col_df_filtered['Vides (%)'] < 80]['Colonne'].tolist()
cols_to_keep_new = [c for c in cols_to_keep_new if c not in cols_not_useful + cols_derived]

for c in cols_to_keep_new:
    pct = col_df_filtered[col_df_filtered['Colonne'] == c]['Vides (%)'].values[0]
    print(f"   ‚úì {c:<35} ({pct}% vide)")

print(f"\nüìä Total colonnes √† garder : {len(cols_to_keep_new)}")
```
